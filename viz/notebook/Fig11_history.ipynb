```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
```

```python
data_path = "../data/fig11.npz"
xlabel = "Log (base 2) Scaled History Window Length T"
ylabel = "Average Top-5 Loss"
labels = ['DQN', 'MDP']
markersize = 15
```

```python
# Create the plot
plt.rc("font", family="DejaVu Sans")
plt.rcParams['figure.figsize'] = (15, 7)
fig, _ = plt.subplots()
ax = plt.subplot()

# Configure the subplot setting
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=20)
ax.yaxis.grid(True, color='#C0C0C0')
ax.xaxis.grid(True, color='#C0C0C0')
ax.spines['top'].set_color('#606060')
ax.spines['bottom'].set_color('#606060')
ax.spines['left'].set_color('#606060')
ax.spines['right'].set_color('#606060')
ax.set_xlabel(xlabel, labelpad=25, color='#333333', size=30)
ax.set_ylabel(ylabel, labelpad=25, color='#333333', size=30)

# Retrieve data from npz file
npz_data = np.load(data_path)
xdata = np.arange(1, 8)

# Plot the simulation data
ax.plot(xdata, npz_data['dqn'], linewidth=3, color='blue', marker='^', markersize=markersize)
ax.plot(xdata, npz_data['mdp']*np.ones_like(xdata), linewidth=3, color='red')
lines = list()
lines.append(Line2D([0], [0], color='blue', linewidth=3, marker='^', markersize=12))
lines.append(Line2D([0], [0], color='red', linewidth=3))
legend = plt.legend(lines, labels, loc='center', bbox_to_anchor=(0.5, -0.32), prop={'size': 30}, ncol=2, frameon=False)
plt.tight_layout()
plt.savefig('../pdf/Fig11.pdf', bbox_inches='tight')
plt.show()
```
