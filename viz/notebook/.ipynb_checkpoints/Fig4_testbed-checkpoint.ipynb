```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import numpy as np
```

```python
data_path = "../data/fig4.npz"
xlabel = "Image Arrival Time (s)"
ylabels = ["Offloading Metric", "Token Count", "Time (ms)"]
color0 = ['#0000ff', '#bfbfbf']
color1 = ['#0000ff', '#ff0000']
color2 = ['#ff0000', '#0000ff', '#ff9900', '#33cc33']
label0 = ['offloaded', 'not offloaded']
label1 = ['offloaded', 'not offloaded']
label2 = ['weak', 'dqn', 'transmission', 'strong']
start, end = 50, 150
```

```python
# Create the plot
plt.rc("font", family="DejaVu Sans")
plt.rcParams['figure.figsize'] = (30, 20)
fig, _ = plt.subplots()
ax0 = plt.subplot(311)
ax1, ax2 = plt.subplot(312, sharex=ax0), plt.subplot(313, sharex=ax0)
axes = [ax0, ax1, ax2]

# Configure the subplot setting
for ax, ylabel in zip(axes, ylabels):
    ax.tick_params(axis='x', labelsize=30)
    ax.tick_params(axis='y', labelsize=30)
    ax.yaxis.grid(True, color='#C0C0C0')
    ax.xaxis.grid(True, color='#C0C0C0')
    ax.spines['top'].set_color('#606060')
    ax.spines['bottom'].set_color('#606060')
    ax.spines['left'].set_color('#606060')
    ax.spines['right'].set_color('#606060')
    ax.set_ylabel(ylabel, labelpad=25, color='#333333', size=40)
ax2.set_xlabel(xlabel, labelpad=20, color='#333333', size=40)
fig.align_ylabels(axes)

# Retrieve data from npz file
data = np.load(data_path)
xdata = data['arrival'][start:end]
dqn_r = data['offload'][start:end]
metric_r = data['metric'][start:end]
token = data['token'][start:end]
delay = data['delay'][:, start:end]
weak, dqn, trans, strong = delay
xdata -= xdata[0]
trace = 1 - dqn_r

# Plot the first subplot
color = [color0[t] for t in trace]
ax0.scatter(xdata, metric_r, color=color, s=80)
dots = []
for i in range(2):
    dots.append(Line2D([0], [0], marker='o', color='w', markerfacecolor=color0[i], markersize=20))
legend1 = fig.legend(dots, label0, loc='upper right', bbox_to_anchor=(0.99, 0.98), prop={'size': 25}, ncol=1, frameon=True)

# Plot the second subplot
for i in range(end - start - 1):
    color = color1[0] if dqn_r[i] else color1[1]
    ax1.plot([xdata[i], xdata[i + 1]], [token[i], token[i + 1]], color=color, linewidth=3, marker='o', markersize=10)
dots = []
for i in range(2):
    dots.append(Line2D([0], [0], marker='o', color=color1[i], markerfacecolor=color1[i], markersize=20, linewidth=6))
legend1 = fig.legend(dots, label1, loc='upper right', bbox_to_anchor=(0.99, 0.662), prop={'size': 25}, ncol=1, frameon=True)

# Plot the third subplot
bar_width = 0.15
xdata += bar_width / 2
y_offset = np.zeros((end - start))
# Configure the zoomed-in sub-region
axins = zoomed_inset_axes(ax2, 5, loc='upper right', bbox_to_anchor=(0.62, 0.7), bbox_transform=ax2.transAxes)
x1, x2, y1, y2 = 12.375, 13.375, 19, 23
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.yaxis.grid(True, color='#C0C0C0')
axins.xaxis.grid(True, color='#C0C0C0')
axins.tick_params(axis='x', bottom=False, top=False, labelsize=18)
axins.tick_params(axis='y', left=False, right=False, labelsize=18)
spines = ['top', 'bottom', 'left', 'right']
for spine in spines:
    axins.spines[spine].set_color('#606060')
    axins.spines[spine].set_linewidth(2.5)
for idx, delay in enumerate([weak, dqn, trans, strong]):
    delay *= 1000
    ax2.bar(xdata, delay, bar_width, bottom=y_offset, color=color2[idx])
    axins.bar(xdata, delay, bar_width, bottom=y_offset, color=color2[idx])
    y_offset = y_offset + delay

# draw a bbox (in the main plot) and connect it to the zoomed-in area
mark_inset(ax2, axins, loc1=3, loc2=1, fc="none", ec='#606060', lw=2.5)
recs = []
for i in range(3, -1, -1):
    recs.append(plt.Rectangle((0, 0), 1, 1, color=color2[i]))
label2.reverse()
legend1 = fig.legend(recs, label2, loc='upper right', bbox_to_anchor=(0.99, 0.335), prop={'size': 25}, ncol=1,
                     frameon=True)

# Save the figure
plt.tight_layout()
plt.savefig('../pdf/Fig4_testbed.pdf', bbox_inches='tight')
plt.show()
```

```python
weak, dqn, trans, strong = data['delay'] * 1000
ofld = data['offload'].astype(bool)
print("Average time (ms) spent on each component of the image classification pipeline:")
print(f"Weak Classifier: {np.mean(weak):.2f} ({np.std(weak):.2f})")
print(f"DQN: {np.mean(dqn):.2f} ({np.std(dqn):.2f})")
print(f"Transmission: {np.mean(trans[ofld]):.2f} ({np.std(trans[ofld]):.2f})")
print(f"Strong Classifier: {np.mean(strong[ofld]):.2f} ({np.std(strong[ofld]):.2f})")
```

```python
nofld = np.logical_not(ofld)
print("Percentage of time spent on each component for offloaded images:")
total = np.mean((weak + dqn + trans + strong)[ofld])
print(f"Weak Classifier: {(np.mean(weak[ofld])/total)*100:.2f}%")
print(f"DQN: {(np.mean(dqn[ofld])/total)*100:.2f}%")
print(f"Transmission: {(np.mean(trans[ofld])/total)*100:.2f}%")
print(f"Strong Classifier: {(np.mean(strong[ofld])/total)*100:.2f}%")
print("Percentage of time spent on each component for local classified images:")
total = np.mean((weak + dqn)[nofld])
print(f"Weak Classifier: {(np.mean(weak[nofld])/total)*100:.2f}%")
print(f"DQN: {(np.mean(dqn[nofld])/total)*100:.2f}%")
```
